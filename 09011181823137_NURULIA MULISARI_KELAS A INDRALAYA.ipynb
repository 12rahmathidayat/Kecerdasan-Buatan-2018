# B. ARTIFICIAL NEURAL NETWORKS (ANN)
Jaringan syaraf tiruan (ANN) amerupakan sistem pemrosesan informasi yang mempunyai karakteristik mirip dengan jaringan syaraf biologi, yang merupakan representasi tiruan dari otak manusia, yang berisikan berjuta-juta sel syaraf (neuron) dan berfungsi untuk memproses informasi. neuron mempunyai karakteristik yang sama dalam ANN, terdiri dalam kelompok-kelompok yang disebut dengan layer. Neuron-neuron dalam satu layer terhubung dalam layer-layer lainnya yang berdekatan. Kekuatan hubungan antar neuron yang berdekatan direpresentasikan dalam kekuatan hubungan atau bobot. ANN pada dasarnya mempunyai 3 layer, antara lain : input layer (terdiri dari neuron yang menerima sebuah input dari lingkungan luar input yang dimasukkan pun merupakan penggambaran dari suatu masalah), hidden layer (terdiri dari neuron-neuron yang menerima masukkan dari input layer kemudian kemudian membawa output pada layer berikutnya) dan output layer (terdiri dari neuron-neuron yang menerima output dari hidden layer dan mengirimkannya kepada pemakai).
Keuntungan dari penggunaan ANN adalah kemampuannya untuk mempelajari hubungan yang tidak diketahui yang sudah ada sebelumnya anatara data input dan output dari setiap sistem. Selain itu pemodelan ANN mempunyai atribut yang diinginkan dan kemampuan belajar dari contoh-contoh tanpa memerlukan data fisik secara eksplisit.

# H. SPLITTING THE DATASET INTO TRAINING AND TEST DATA
Membagi kumpulan data yang telah diproses menjadi data pelatihan dan juga Pengujian. Ukuran data uji diambil menjadi 15% dari keseluruhan data (yang berisikan 115 observasi) dan model akan dilatih pada 653 observasi.

```python
X = df.iloc[:,0:-1].values
y = df.iloc[:,-1].values
```

```python
X_train,X_test,y_train,y_test = train_test_split(X,y,test_size=0.2)
```

* Ketidakseimbangan yang sangat besar dapat diamati pada nilai y_train. Untuk mengatasi masalah tersebut bisa digunakan dengan teknik SMOTE. Synthetic Minority Oversampling Technique (SMOTE) digunakan untuk menghilangkan ketidakseimbangan dalam data pelatihan dengan menggunakan data saat ini. [5]SUMBER :https://towardsdatascience.com/pima-indian-diabetes-prediction-7573698bd5fe.  

```python
sc = StandardScaler()
X_train = sc.fit_transform(X_train)
X_test = sc.transform(X_test)
```

```python
import tensorflow as tf
ann = tf.keras.Sequential()
```

* TensorFlow merupakan sebuah framework komputasional untuk membuat model machine learning. TensorFlow menyediakan berbagai toolkit yang memungkinkan kita untuk membuat model pada tingkat abstraksi yang kita sukai. Kita dapat menggunakan API dengan tingkat yang lebih rendah untuk membuat model dengan menentukan serangkaian operasi matematis. Tensorflow dapat membantu kita dalam membuat neural network (jaringan artifisial yang mirip otak manusia) dalam skala besar.

```python
ann.add(tf.keras.layers.Dense(units=12,activation='relu'))
ann.add(tf.keras.layers.Dense(units=12,activation='relu'))
ann.add(tf.keras.layers.Dense(units=12,activation='relu'))
ann.add(tf.keras.layers.Dense(units=1,activation='sigmoid'))
```

```python
ann.compile(optimizer='adam',loss='binary_crossentropy',metrics=['accuracy'])
```

```python
ann.fit(X_train,y_train,batch_size=12,epochs=100)
```

```python
y_pred = ann.predict(X_test)
y_pred = (y_pred>0.75)
```

```python
for i in range(len(y_pred)):
    print("Predicted %d------> Expected %d" %(y_pred[[i]],y_test[i]))
```

```python
from sklearn.metrics import confusion_matrix,accuracy_score
cm = confusion_matrix(y_pred,y_test)
acc = accuracy_score(y_pred,y_test)
```

* DEFINISI BACKPROPAGATION

Backpropagation (propagasi balik) merupakan salah satu jaringan syaraf tiruan (Neural Network) yang merupakan metode pelatihan yang diawasi (Supervised Learning) dengan jaringan multi-layer dan memiliki fitur khusus untuk meminimalkan kesalahan pada keluaran yang dihasilkan oleh jaringan.
Biasanya dalam proses klasifikasi Jaringan Syaraf Tiruan Backpropagation. Pengklasifikasi ini bekerja dengan melakukan dua tahap perhitungan yaitu perhitungan maju yang akan menghitung nilai kesalahan antara nilai keluaran sistem dengan nilai semestinya dan perhitungan mundur untuk mengoreksi bobot berdasarkan nilai kesalahan. [6]SUMBER :https://hendroprasetyo.com/backpropagation-neural-network-jaringan-saraf-tiruan/#.X5a6LIgzbIU
Backpropagation merupakan salah satu model yang terdapat pada JST yang menggunakan supervised learning. Algoritma ini sering digunakan untuk menyelesaikan suatu masalah yang rumit. Hal ini dikarenakan algoritma ini dilatih menggunakan metode
pembelajaran. Pada jaringan ini diberikan pola yang sepasang yang terdiri atas pola masukan dan pola yang dikehendaki (Arizona, 2013). Backpropagation memiliki proses pembelajaran
