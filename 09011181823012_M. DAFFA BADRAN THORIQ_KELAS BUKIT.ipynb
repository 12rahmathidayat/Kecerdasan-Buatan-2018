###  Define Design ANN

Design pada ANN berupa penentuan arsitektur network, menentukan berapa jumlah nilai bobot, berapa layer yang digunakan dalam satu network, serta pada masing masing layer yang digunakan, memiliki berapa jumlah nodes

```python
def initialize_param(layers_with_nodes,in_data,out_data):
    
    1. #input layer's values.
    # jumlah ndoes pada input layer adalah berdasarkan parameter jumlah layers_with_nodes yang di definisikan di atas
    layer_amount = len(layers_with_nodes)
    print(layer_amount)
    
    # Masukkan parameter yang menghubungkan input layer dan hidden layer yaitu adalah weight1 (bobot) dan ditambah bias1 dengan nilai 1
    # Fungsi bias ini agar menghindari seluruh output bernilai 0 yang biasa disebut bias term
    parameters = {}
    # Nilai weight1 dimasukkan nilai random dalam bentuk array dengan bantuan library numpy
    parameters["weight1"] = np.random.randn(layers_with_nodes[0],in_data.shape[0]) * 0.1
    parameters["bias1"] = np.zeros((layers_with_nodes[0],1))
    
    print("" + str(layer_amount+2) + " layers. (" + str(layer_amount) + " hidden layer)")

    
    2. #the hidden layer's values.
    # Untuk masing masing layer, memiliki jumlah bobot yang berbeda beda
    for i in range(layer_amount-1):
        #print(i+2)
        # weight dan bias di set dengan cara yang sama dengan saat input layer tadi, perbedaannya,
        # dikarenakan hidden layer terdiri dari lebih dari 1 layer, maka digunakan perulangan dengan for
        # untuk menentukan weight dan bias.
        w = "weight" + str(i+2)
        b = "bias" + str(i+2)
        parameters[w] = np.random.randn(layers_with_nodes[i+1],layers_with_nodes[i]) * 0.1
        parameters[b] = np.zeros((layers_with_nodes[i+1],1))
    
    3. #output layer's values.
    # Pada output layer, kita tetap memiliki bobot dan bias juga karena setiap layer memiliki bobot dan bias masing masing, sehingga kita perlu mendefinisikan bobot dan bias untuk output layer.
    lastw = "weight" + str(layer_amount+1)
    lastb = "bias" + str(layer_amount+1)
    parameters[lastw] = np.random.randn(out_data.shape[0],layers_with_nodes[layer_amount-1]) * 0.1
    parameters[lastb] = np.zeros((out_data.shape[0],1))
    
    return parameters, (layer_amount+1)
```

### Define Fungsi Aktivasi

##### Sigmoid

Karena kita menggunakan sigmoid, maka rumus sigmoid adalah sebagai berikut :

\begin{align*}
sigmoid = \frac{1} {1 + e^{z}}
\end{align*}

dimana z = hasil output dari sebuah layer, yang didapatkan dengan cara 
\begin{align*}
z = (w1*x1 + w2*x2 + w3*x3 + w4*x4 + ...... + w[n]*x[n])
\end{align*}

Keterangan :
w = bobot, 
x = input

##### Tanh

Untuk tanh activation function, tanh akan mengubah nilai asli ke dalam array range [-1,1]. Tanh adalah non linear, tetapi outputnya zero-centered

Rumus dari tanh function adalah
\begin{align*}
tanh(z) =  \frac{e^{z} - e^{-z}} {e^{z} +e^{-z}}
\end{align*}

dimana z = hasil output dari sebuah layer, yang didapatkan dengan cara 
\begin{align*}
z = (w1*x1 + w2*x2 + w3*x3 + w4*x4 + ...... + w[n]*x[n])
\end{align*}

Keterangan :
w = bobot, 
x = input

##### Define loss 

a. Berikut adalah rumus untuk menghitung loss pada Output layer <br>
$\partial$r= ($t_r$ - $y_r$) . f'($y'_k$) <br>

b Berikut adalah rumus untuk menghitung loss pada Hidden Layer <br>
$\partial$j = $\partial$'j . f'($\partial$'j)<br>

```python
def sigmoid(x):
    y_head = 1/(1 + np.exp(-x))   # basic sigmoid for output layer
    return y_head;

def compute_cost(A_f, y_train):
    logprobs = np.multiply(np.log(A_f),y_train)
    cost = -np.sum(logprobs)/y_train.shape[1]    # computing the loss value
    return cost
```

### Define Feed Forward / Forward Propagation

Pada forward propagation, program ANN akan dimulai dari input layer menuju output layer, dengan argument paramater berupa in_data (data input), parameters yang telah di definisikan pada intialize_params dan connection_amount berupa jumlah connection pada arsitektur ANN tersebut. Pada feed forward, program akan menghitung berapa jumlah error yang didapat ketika memasukkan weight yang awal.

#### Perhitungan

1. Input layer akan bergerak ke hidden layer pertama, 
\begin{align*}
z'j = \sum\limits _{i=1} ^{m} . Xi . Wij
\end{align*}
<br>
\begin{align*}
z'j = (w1[1]*x1 + w2[1]*x2 + w3[1]*x3 + w4[1]*x4 + ...... + w[n]*x[n])
\end{align*}
Keterangan : <br>
X = input <br>
W = Bobot <br>
z = output hidden layer sebelum dimasukkan fungsi aktivasi <br>
i = nodes ke berapa (input) <br>
j = nodes ke berapa (hidden layer 1) <br>
<br><br>

2. Hitung fungsi aktivasi pada layer pertama dengan tanh function activation
\begin{align*}
tanh(z) =  \frac{e^{z} - e^{-z}} {e^{z} +e^{-z}}
\end{align*}
Keterangan : <br>
z = output hidden layer sebelum dimasukkan fungsi aktivasi <br>
<br><br>

3. Dari hasil z tadi, kemudian bergerak dari hidden layer 1 ke hidden layer ke 2, perhitungan masih menggunakan rumus yang sama untuk mencari nilai z'j. 
\begin{align*}
o'n = \sum\limits _{k=1} ^{q} . zk . Wkp
\end{align*}
Keterangan : <br>
z = output dari hidden layer 1 yang telah dimasukkan fungsi aktivasi <br>
k = input nodes ke berapa (hidden layer 1) <br>
w = bobot  <br>
p = output nodes ke berapa (hidden layer 2) <br>
<br>
<br><br>

4. Lanjut hitung nilai output pada Hidden layer 2 dengan function activation tanh tadi, nilai input bukan lagi berasal dari input nodes pada input layer, tetapi berasal dari nilai output yang telah melewati fungsi aktivasi pada hidden layer 1.
\begin{align*}
tanh(o) = \frac{e^{o} - e^{-o}} {e^{o} + e^{-o}}
\end{align*}
Keterangan : <br>
o= output dari hidden layer 1 yang belum dimasukkan fungsi tanh
<br><br>

5. Dari hidden layer ke dua lanjut ke output layer, kita akan menghitung nilai output juga pada output layer dengan rumus yang sama seperti sebelumnya dengan mencari nilai output pada output layer 
\begin{align*}
y'r = \sum\limits _{j = 1} ^{p} . oj . wjh
\end{align*}
Keterangan : <br>
o = output dari hidden layer 2 yang telah di masukkan fungsi aktivasi <br>
j = input nodes ke berapa <br>
h = output nodes ke berapa <br>
w = bobot dari hidden layer 2 menuju output layer <br>
<br><br>

6. Hitung nilai output pada output layer dengan menggunakan fungsi aktivasi, Untuk bagian output layer, kita menggunakan fugnsi aktivasi sigmoid dengan rumus 
\begin{align*}
sigmoid = \frac{1} {1 + e^{y}}
\end{align*}
Keterangan : <br>
y = hasil output dari output layer dari perhitungan bobot * input sebelum dimasukkan ke dalam fungsi aktivasi<br>
<br><br>

7. Pada program feed forward, kita juga ahrus menghitung nilai loss, nilai loss juga berada pada setiap layer, baik hidden layer maupun output layer.<br>
a. Berikut adalah rumus untuk menghitung loss pada Output layer <br>
$\partial$r= ($t_r$ - $y_r$) . f'($y'_k$) <br>

b Berikut adalah rumus untuk menghitung loss pada Hidden Layer <br>
$\partial$j = $\partial$'j . f'($\partial$'j)<br>

```python
def forw_prop_NN(in_data, parameters, connection_amount): # forward propagation
    
    cache = {}
    # Data input di inisiasikan sebagai a0
    a0 = in_data
    
    
    # Hidden layer
    # Karena memiliki lebih dari satu hidden layer, maka dilakukanlah perulangan (looping) 
    # Looping digunakan untuk menghitung nilai output pada masing masing hidden layer
    for i in range(connection_amount-1):
        w = "weight" + str(i+1)   
        b = "bias" + str(i+1) 
        
        z = "Z" + str(i+1)
        a = "A" + str(i+1)
        
        # Karena hidden layer menggunakan activation function tanh, maka kita akan menghitung output 
        # dengan menggunakan activation function tanh, untuk z sebagai output dari bobot * input nodes,
        # sementara untuk a adalah output dari hidden layer yang telah diberikan fungsi aktivasi dengan parameter 
        # hasil dari nilai z tadi dari perhitungan z = (w1*x1 + w2*x2 + .... + w187 * x187)
        cache[z] = np.dot(parameters[w],a0) + parameters[b]
        cache[a] = np.tanh(cache[z])
        a0 = cache[a]
        
        
    # output layer
    
    z_fin = "Z" + str(connection_amount)
    a_fin = "A" + str(connection_amount)
    
    w = "weight" + str(connection_amount)   
    b = "bias" + str(connection_amount)
    
    a0 = "A" + str(connection_amount-1)
    
    # Pada output layer kita mencari nilai output dengan cara yang sama
    # dengan menggunakan z_fin = (w*x) dan ditambahkan bias diujung
    # Untuk fungsi aktivasinya berbeda dengan saat di hidden layer, jika di output layer menggunakan fungsi aktivasi sigmoid
    # Fungsi sigmoid telah di definisikan sebelumnya sehingga dapat dipanggil ke dalam code.
    cache[z_fin] = np.dot(parameters[w],cache[a0]) + parameters[b]
    cache[a_fin] = sigmoid(cache[z_fin])
    A = cache[a_fin]
 
    # Return dan kembalikan nilai A yaitu adlah nilai output yang telah dimasukkan fungsi aktivasi sigmoid
    return A, cache
```

### Define Back Propagation 

Pada backpropagation, program akan bergerak mundur dari output menuju input layer dengan mundur satu persatu layer terlebih dahulu, fungsi dari abck propagation ini adalah untuk mengubah nilai weight (bobot) sebelumnya pada masing masing layer yang bertujuan agar jumlah error yang berada pada output semakin mengecil. 

#### Perhitungan

1. Lakukan perhitungan terahdap nilai error pada output layer dengan mengurangi nilai output yang telah melewati fungsi aktivasi sigmoid dengan target output yang diinginkan pada y_train

2.  Update nilai weight dan nilai bias yang menghubungkan hidden layer 2 dan layer output <br>
$W_rj$ = $W_rj$ + $\alpha$ $\partial$k . o'k <br>
Dimana : <br>
o'k = (Xkp . Wkp + ..... + Xkp . Wkp) <br>
Misal <br>
o'1 = (X1[1] . W1[1] + X2[1] . W2[1] + ........... + X187[1] . W187[1]) <br>
....................................................................... <br>
o'8 = (X1[8] . W1[8] + X2[8] . W2[8] + ........... + X187[8] . W187[8]) <br>
<br><br>

3. Lalu hitung output pada masing masing layer, bandingkan dengan nilai error, berapa banyak kontribusi mereka dalam error program. Jika sudah makan lakukan perhitungan weight seperti diatas, karena hidden layer memiliki 2 layer, jadi bisa menggunakan perulangan untuk mempermudah codingan.

4. Setelah sampai pada input layer, maka itu ada weight terakhir yang perlu dirubah. Dengan melakukan satu kali feed forward dan satu kali back propagation, maka program ANN telah menyelesaikan satu Epoh (1 Iterasi). Program akan kembali melakukan hal yang sama sesuai dengan set iterasi yang diberikan, atau sampai error mengecil atau telah convergence

```python
def backw_prop_NN(parameters,cache,x_train,y_train,connection_amount):  #backpropagation
    
    dz = {}
    grads = {}
    
    
    # firstly start with output layer.
    
    a_fin = "A" + str(connection_amount)
    a_pre_fin = "A" + str(connection_amount-1)
    w_fin = "dweight" + str(connection_amount)
    b_fin = "dbias" + str(connection_amount)
    
    # Kita hitung berapa error yang kita dapatkan, caranya dengan mengurangi nilai a_fin dengan nilai y_train
    # a_fin adalah nilai output yang telah dimasukkan fungsi aktivasi sigmoid, dan y_train adalah hasil training data di awal
    dz[connection_amount] = cache[a_fin] - y_train
    
    # Hitung nilai weight dan nilai bias yang baru untuk hidden layer ke2 menuju output layer
    grads[w_fin] = np.dot(dz[connection_amount],cache[a_pre_fin].T)/x_train.shape[1]
    grads[b_fin] = np.sum(dz[connection_amount], axis=1, keepdims=True)/x_train.shape[1]
    
    # other layers.
    # Untuk hidden layer 1 ke hidden layer 2 dan juga dari input layer menuju hidden layer 1
    for i in range((connection_amount-1),1,-1):
        a = "A" + str(i)
        a_pre = "A" + str(i-1)
        dw = "dweight" + str(i)
        db = "dbias" + str(i)
        w = "weight" + str(i+1)
        
        dz[i] = np.dot(parameters[w].T,dz[i+1])*(1 - np.power(cache[a], 2))
        grads[dw] = np.dot(dz[i],cache[a_pre].T)/x_train.shape[1]
        grads[db] = np.sum(dz[i], axis=1,keepdims=True)/x_train.shape[1]
    
    dz[1] = np.dot(parameters["weight2"].T,dz[2])*(1 - np.power(cache["A1"], 2))
    grads["dweight1"] = np.dot(dz[1],x_train.T)/x_train.shape[1]
    grads["dbias1"] = np.sum(dz[1],axis =1,keepdims=True)/x_train.shape[1]
    
    return grads
```

#### Ubah seluruh nilai parameter (Weight dan Bias) untuk iterasi selanjutnya

Setelah melakukan perubahan pada seluruh weight pada setiap layer, maka kita akan masuk ke Epoh (iterasi selanjtunya) dengan parameter baru yang berupa weight dan bias yang baru dan akan melakukan feed forward untuk Epoh ke 2

```python
def update_param(parameters, grads, connection_amount, lr=0.01): #update
    for i in range(connection_amount):
        w = "weight" + str(i+1)
        b = "bias" + str(i+1)
        dw = "dweight" + str(i+1)
        db = "dbias" + str(i+1)
        
        parameters[w] = parameters[w] - lr*grads[dw]
        parameters[b] = parameters[b] - lr*grads[db]
        
    return parameters                
```

#### Lakukan prediction dengan mengumpulkan seluruh nilai output pada masing masing iterasi

```python
def predict(A,parameters,x_test,connection_amount): #prediction
    
    A, cache = forw_prop_NN(x_test,parameters,connection_amount)
    Y_prediction = np.zeros((1,x_test.shape[1]))
 
    for i in range(A.shape[1]):
        if A[0,i]<= 0.5:
            Y_prediction[0,i] = 0
        else:
            Y_prediction[0,i] = 1

    return Y_prediction             # basic binary prediction.
```

## Jalankan Program ANN

Setelah mendefinsikan seluruh aspek yang diperlukan dalam program, seperti define input, jumlah layer, jumlah nodes tiap layer, define fungsi aktivasi yang akan digunakan, serta melakukan define untuk memberitahu program apa saja yang akan dilakukan saat feed froward dan back propagation, sekarang kita jalankan program ANN dengan input yang kita miliki. Program menset iterasi sebanyak 2800 kali sesuai jumlah num_iter, lalu menjalankan program feed forward dan back propagation. Lalu kita kan mengecek perubahan error per 100 kali iterasi, apakah ada penurunan atau tidak

```python
def multi_layer_model(layers_with_nodes,x_train,y_train,x_test,y_test, num_iter, lr = 0.01):    
    
    cost_list = []     
    index_list = []
    
    parameters, connection_amount = initialize_param(layers_with_nodes, x_train, y_train)   
    
    # for each iteration..
    # Perulangan sebanyak 2800 kali 
    for i in range(0, num_iter):
         # forward propagation Program dijalankan
        A, cache = forw_prop_NN(x_train,parameters,connection_amount)
        # compute cost ( Hitung loss untuk setiap layer (hidden dan output))
        cost = compute_cost(A, y_train)
         # backward propagation ( Setelah mendapat error, lakukan abck propagation dan update nilai weight dan bias)
        grads = backw_prop_NN(parameters,cache,x_train,y_train,connection_amount)
         # update parameters
        parameters = update_param(parameters, grads, connection_amount, lr)
        # Tampilkan nilai error setiap 100 kali perulangan untuk melihat statistik penurunan error
        if i % 100 == 0:
            cost_list.append(cost)
            index_list.append(i)
            print ("Cost after iteration %i: %f" %(i, cost))

    # Tampilkan grafik 2 dimensi dengan sumbu x Number of iteration dan y adalah cost (loss) nya
    # jumlah iterasi diinisiasi dengan index_list dan untuk loss nya diinisiasi dengan cost_list
    plt.plot(index_list,cost_list)
    plt.xticks(index_list,rotation='vertical')
    plt.xlabel("Number of Iterarion")
    plt.ylabel("Cost")
    plt.show()
    
    # predict
    
    y_prediction_train = predict(A, parameters, x_train, connection_amount)
    y_prediction_test = predict(A, parameters, x_test, connection_amount)
    
    
    # Print train/test Errors
    print("train accuracy: {} %".format(100 - np.mean(np.abs(y_prediction_train - y_train)) * 100))
    print("test accuracy: {} %".format(100 - np.mean(np.abs(y_prediction_test - y_test)) * 100))
    return parameters, connection_amount
```

```python
parameters, connection_amount=multi_layer_model([16,8],x_train,y_train,x_test,y_test,num_iter=2800,lr=0.3)

# the first array is where we desing our net for hidden layers with nodes.
```

### Kesimpulan

Terlihat dari cost after iteration, loss (error) berkurang setiap perulangan (Epoh/Iterasi), yang menandakan bahwa program berjalan dengan baik, karena normalnya setiap iterasi, nilai loss akan berubah menjadi lebih kecil hingga 0 (tidak ada loss lagi). Selain itu juga akurasi training 90% dan test akurasi 91% yang memiliki nilai sangat baik.
