## SVM

SVM merupakan kepanjangan dari Support Vector Machine yang pada dasarnya berupa algoritma klasifikasi untuk data linear dan non-linear. SVM biasanya menggunakan mapping non-linear untuk mentransformasi data awal ke dimensi yang lebih tinggi.

Kali ini kita akan melakukan koding sederhana dari bentuk SVM.

Berikut kode awalnya:

```python
from sklearn.datasets.samples_generator import make_blobs
from matplotlib import pyplot as plt
import numpy as np
import pandas as pd
from sklearn.model_selection import train_test_split
from sklearn.metrics import confusion_matrix
from sklearn.svm import SVC
```

Setelah dilakukan pembuatan kode awalnya, kita akan langsung melakukan koding untuk memisahkan antar 2 bagian dari sebuah data non-linear:

```python
X, y = make_blobs(n_samples=125, centers=2, cluster_std=0.60, random_state=0)

train_X, test_X, train_y, test_y = train_test_split(X, y, test_size=20, random_state=0)

plt.scatter(train_X[:, 0], train_X[:, 1], c=train_y, cmap='summer' )
```

Setelah 2 data terbuat, selanjutnya kita akan membuat kernel menjadi linear untuk mengisi train test:

```python
svc = SVC(kernel='linear')
svc.fit(train_X, train_y)
```

Setelah SVC sudah ditentukan, saatnya membagi 2 data tersebut:

```python
plt.scatter(train_X[:, 0], train_X[:, 1], c=train_y, cmap='summer' )

ax = plt.gca()
xlim = ax.get_xlim()

ax.scatter(test_X[:, 0], test_X[:, 1], c=test_y, cmap='summer', marker='s' )

w = svc.coef_[0]
a = -w[0]/w[1]
xx = np.linspace(xlim[0], xlim[1])
yy = a * xx - (svc.intercept_[0] / w[1])
plt.plot(xx,yy)
plt.show()
```

Untuk memprediksi matriks confusion, dilakukanlah koding dibawah ini:

```python
pred_y = svc.predict(test_X)
```

```python
confusion_matrix(test_y, pred_y)
```

Setelah dilakukan perhitungan, bisa kita lihat terdapat 2 data yang telah terpisah dan telah diklasifikasi dalam SVM yang terlampir

Kesimpulan yang bisa kita ambil disini adalah kita bisa mengetahui jikalau data yang tadinya tercampur kini sudah terpisah sesuai dengan kelasnya masing"
