## 1.1 Apa itu Support Vector Machine?

Tujuan dari algoritma support vector machine adalah untuk menemukan hyperplane dalam ruang dimensi-N (N - jumlah fitur) yang secara jelas mengklasifikasikan titik-titik data.

Untuk memisahkan kedua kelas titik data tersebut, ada banyak kemungkinan hyperplanes yang bisa dipilih. Tujuan kita adalah menemukan bidang yang memiliki margin maksimum, yaitu jarak maksimum antara titik data kedua kelas. Memaksimalkan jarak margin memberikan beberapa penguatan sehingga poin data di masa mendatang dapat diklasifikasikan dengan lebih yakin.

## 1.2 Hyperplanes and Support Vectors

Hyperplanes adalah batas keputusan yang membantu mengklasifikasikan titik data. Titik data yang berada di kedua sisi hyperplane dapat dikaitkan dengan kelas yang berbeda. Selain itu, dimensi hyperplane tergantung pada jumlah fitur. Jika jumlah fitur inputnya 2, maka hyperplane hanyalah sebuah garis. Jika jumlah fitur masukan adalah 3, maka hyperplane menjadi bidang dua dimensi. Sulit dibayangkan bila jumlah fitur melebihi 3.

Support vector machine adalah titik-titik data yang lebih dekat ke hyperplane dan mempengaruhi posisi dan orientasi hyperplane. Dengan menggunakan support machine ini, kami memaksimalkan margin pengklasifikasi. Menghapus support machine akan mengubah posisi hyperplane. Ini adalah poin yang membantu kami membangun SVM kami.

## 1.3 Intuisi Margin Besar

Dalam regresi logistik, kami mengambil output dari fungsi linier dan menekan nilai dalam kisaran [0,1] menggunakan fungsi sigmoid. Jika nilai squashed lebih besar dari nilai ambang (0,5) kami memberinya label 1, jika tidak kami memberikan label 0. Dalam SVM, kami mengambil output dari fungsi linier dan jika output lebih besar dari 1, kami mengidentifikasi itu dengan satu kelas dan jika outputnya -1, kami mengidentifikasi dengan kelas lain. Karena nilai ambang batas diubah menjadi 1 dan -1 dalam SVM, kami memperoleh rentang nilai penguatan ini ([- 1,1]) yang bertindak sebagai margin.

## 2.2 Model Training (Menemukan solusi dari masalah)

```python
from sklearn.model_selection import cross_val_score, train_test_split
from sklearn.pipeline import Pipeline
from sklearn.preprocessing import StandardScaler, MinMaxScaler

X = df.drop('target', axis=1)
y = df.target

print(f"'X' shape: {X.shape}")
print(f"'y' shape: {y.shape}")

pipeline = Pipeline([
    ('min_max_scaler', MinMaxScaler()),
    ('std_scaler', StandardScaler())
])

X_train, X_test, y_train, y_test = train_test_split(X, y, test_size=0.3, random_state=42)
```

```python
from sklearn.metrics import accuracy_score, confusion_matrix, classification_report

def print_score(clf, X_train, y_train, X_test, y_test, train=True):
    if train:
        pred = clf.predict(X_train)
        clf_report = pd.DataFrame(classification_report(y_train, pred, output_dict=True))
        print("Train Result:\n================================================")
        print(f"Accuracy Score: {accuracy_score(y_train, pred) * 100:.2f}%")
        print("_______________________________________________")
        print(f"CLASSIFICATION REPORT:\n{clf_report}")
        print("_______________________________________________")
        print(f"Confusion Matrix: \n {confusion_matrix(y_train, pred)}\n")
        
    elif train==False:
        pred = clf.predict(X_test)
        clf_report = pd.DataFrame(classification_report(y_test, pred, output_dict=True))
        print("Test Result:\n================================================")        
        print(f"Accuracy Score: {accuracy_score(y_test, pred) * 100:.2f}%")
        print("_______________________________________________")
        print(f"CLASSIFICATION REPORT:\n{clf_report}")
        print("_______________________________________________")
        print(f"Confusion Matrix: \n {confusion_matrix(y_test, pred)}\n")
```

## 2.3 Support Vector Machines (Kernels)

- Parameter C: Mengontrol trade-off antara mengklasifikasikan poin training dengan benar dan memiliki batasan keputusan yang mulus.
    - C kecil (longgar) membuat biaya (penalti) kesalahan klasifikasi rendah (margin lunak)
    - C besar (ketat) membuat biaya kesalahan klasifikasi menjadi tinggi (margin keras), memaksa model untuk menjelaskan data input lebih ketat dan berpotensi melebihi itu.
- Parameter gamma: Mengontrol sejauh mana pengaruh satu set training tercapai.
    - Gamma besar: jangkauan dekat (titik data yang lebih dekat memiliki bobot yang tinggi)
    - Gamma kecil: jangkauan jauh (solusi yang lebih umum)
- parameter derajat: Derajat fungsi kernel polinomial ('poli'). Diabaikan oleh semua kernel lainnya.

Pendekatan umum untuk menemukan nilai hyperparameter yang tepat adalah dengan menggunakan pencarian grid. Seringkali lebih cepat untuk terlebih dahulu melakukan pencarian grid yang sangat kasar, kemudian pencarian grid yang lebih halus di sekitar nilai terbaik yang ditemukan. Memiliki pemahaman yang baik tentang apa yang sebenarnya dilakukan setiap hyperparameter juga dapat membantu Anda mencari di bagian kanan ruang hyperparameter.

### 2.3.1 Linear Kernel SVM

```python
from sklearn.svm import LinearSVC

model = LinearSVC(loss='hinge', dual=True)
model.fit(X_train, y_train)

print_score(model, X_train, y_train, X_test, y_test, train=True)
print_score(model, X_train, y_train, X_test, y_test, train=False)
```

### 2. 3. 2. Polynomial Kernel SVM

Kode ini melatih pengklasifikasi SVM menggunakan kernel ploynomial derajat ke-2.

```python
from sklearn.svm import SVC

# The hyperparameter coef0 controls how much the model is influenced by high degree ploynomials 
model = SVC(kernel='poly', degree=2, gamma='auto', coef0=1, C=5)
model.fit(X_train, y_train)

print_score(model, X_train, y_train, X_test, y_test, train=True)
print_score(model, X_train, y_train, X_test, y_test, train=False)
```

### 2. 3. 3. Radial Kernel SVM

```python
model = SVC(kernel='rbf', gamma=0.5, C=0.1)
model.fit(X_train, y_train)

print_score(model, X_train, y_train, X_test, y_test, train=True)
print_score(model, X_train, y_train, X_test, y_test, train=False)
```

Kernel lain ada tetapi tidak lebih jarang digunakan. Misalnya, beberapa kernel dikhususkan untuk struktur data tertentu. Kernel string terkadang digunakan saat mengklasifikasikan dokumen teks pada urutan DNA.

Dengan begitu banyak kernel untuk dipilih, bagaimana Anda bisa memutuskan mana yang akan digunakan? Sebagai aturan praktis, Anda harus selalu mencoba kernel linier terlebih dahulu, terutama jika set pelatihan sangat besar atau jika memiliki banyak fitur. Jika ukuran pelatihan tidak terlalu besar, Anda harus mencoba kernel RBF Gaussian juga.

## 2. 4. Data Preparation for SVM

Bagian ini mencantumkan beberapa saran tentang cara terbaik mempersiapkan data training Anda saat mempelajari model SVM.

- Input Numerik: SVM mengasumsikan bahwa input Anda adalah numerik. Jika Anda memiliki masukan kategorikal, Anda mungkin perlu mengubahnya menjadi variabel dummy biner (satu variabel untuk setiap kategori).
- Klasifikasi Biner: SVM dasar seperti yang dijelaskan dalam posting ini ditujukan untuk masalah klasifikasi biner (dua kelas). Meskipun, ekstensi telah dikembangkan untuk regresi dan klasifikasi kelas jamak.

```python
X_train = pipeline.fit_transform(X_train)
X_test = pipeline.transform(X_test)
```

```python
print("=======================Linear Kernel SVM==========================")
model = SVC(kernel='linear')
model.fit(X_train, y_train)

print_score(model, X_train, y_train, X_test, y_test, train=True)
print_score(model, X_train, y_train, X_test, y_test, train=False)

print("=======================Polynomial Kernel SVM==========================")
from sklearn.svm import SVC

model = SVC(kernel='poly', degree=2, gamma='auto')
model.fit(X_train, y_train)

print_score(model, X_train, y_train, X_test, y_test, train=True)
print_score(model, X_train, y_train, X_test, y_test, train=False)

print("=======================Radial Kernel SVM==========================")
from sklearn.svm import SVC

model = SVC(kernel='rbf', gamma=1)
model.fit(X_train, y_train)

print_score(model, X_train, y_train, X_test, y_test, train=True)
print_score(model, X_train, y_train, X_test, y_test, train=False)
```

# 3. Support Vector Machine Hyperparameter tuning

```python
from sklearn.model_selection import GridSearchCV

param_grid = {'C': [0.01, 0.1, 0.5, 1, 10, 100], 
              'gamma': [1, 0.75, 0.5, 0.25, 0.1, 0.01, 0.001], 
              'kernel': ['rbf', 'poly', 'linear']} 

grid = GridSearchCV(SVC(), param_grid, refit=True, verbose=1, cv=5, iid=True)
grid.fit(X_train, y_train)

best_params = grid.best_params_
print(f"Best params: {best_params}")

svm_clf = SVC(**best_params)
svm_clf.fit(X_train, y_train)
print_score(svm_clf, X_train, y_train, X_test, y_test, train=True)
print_score(svm_clf, X_train, y_train, X_test, y_test, train=False)
```

## 4. 1. PCA Visualization

Seperti yang telah kita perhatikan sebelumnya, sulit untuk memvisualisasikan data berdimensi tinggi, kita dapat menggunakan PCA untuk menemukan dua komponen utama pertama, dan memvisualisasikan data dalam ruang dua dimensi yang baru ini, dengan single scatter-plot. Sebelum kita melakukan ini, kita perlu menskalakan data kita sehingga setiap fitur memiliki satu unit varian.

```python
scaler = StandardScaler()
X_train, X_test, y_train, y_test = train_test_split(X, y, test_size=0.3, random_state=42)

X_train = scaler.fit_transform(X_train)
X_test = scaler.transform(X_test)
```

PCA dengan Scikit Learn menggunakan proses yang sangat mirip dengan fungsi preprocessing lainnya yang disertakan dengan SciKit Learn. Kami membuat instance objek PCA, menemukan komponen utama menggunakan metode fit, kemudian menerapkan rotasi dan pengurangan dimensi dengan memanggil transform().

Kami juga dapat menentukan berapa banyak komponen yang ingin kami simpan saat membuat objek PCA.

```python
from sklearn.decomposition import PCA

pca = PCA(n_components=3)

X_train = pca.fit_transform(X_train)
X_test = pca.transform(X_test)
```

```python
plt.figure(figsize=(8,6))
plt.scatter(X_train[:,0],X_train[:,1],c=y_train,cmap='plasma')
plt.xlabel('First principal component')
plt.ylabel('Second Principal Component')
```

Jelas dengan menggunakan dua komponen ini kita dapat dengan mudah memisahkan kedua kelas ini.
