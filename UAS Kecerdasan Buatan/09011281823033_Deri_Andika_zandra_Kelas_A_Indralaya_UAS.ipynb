**Support Vector Machine (SVM)**

SVM merupakan suatu teknik untuk melakukan prediksi, baik prediksi dalam kasus regresi maupun klasifikasi. Teknik SVM digunakan untuk mendapatkan fungsi pemisah (hyperplane) yang optimal untuk memisahkan observasi yang memiliki nilai variabel target yang berbeda. Hyperplane ini dapat berupa line pada two dimension dan dapat berupa flat plane pada multiple dimension.

**Mengimpor liblary yang dibutuhkan**

```python
import numpy as np
import matplotlib.pyplot as plt
import pandas as pd
```

**Memasukkan data sheat yang di gunakan**

```python
data = pd.read_csv('Social_Network_Ads.csv')
```

```python
data.head()
```

**mengimport liblary yang di butuhkan**

```python
from sklearn.preprocessing import StandardScaler, LabelEncoder
```

```python
from sklearn.model_selection import train_test_split
```

**Mengubah label data gender menjadi 1 dan 0 menggunakan label encoder**


Apabila male = 1
Apabila Famale = 0



```python
gender_enc = LabelEncoder()
```

```python
data['Gender'] = gender_enc.fit_transform(data['Gender'])
```

Memanggil data

```python
data.head()
```

```python
X = data.iloc[:, :-1]
```

```python
X.head()
```

Menghapus User Id karna data tidak digunakan

```python
X = X.drop('User ID', axis=1)
```

Memanggil data

```python
X.head()
```

**Membuat variabel X dan Y**

```python
y = data.iloc[:, -1].values
y
```

```python
sc = StandardScaler()
```

```python
X = sc.fit_transform(X)
```

**Memasukkan data sheat kedalam classifier**

```python
X_train, X_test, y_train, y_test = train_test_split(X, y, test_size=0.2, random_state=100)
```

```python
from sklearn.svm import SVC
```

**Menggunakan SVM untuk mengklarifikasi data menggunakan classifier**

```python
classifier = SVC(kernel='rbf', random_state=0)
```

```python
classifier.fit(X_train, y_train)
```

**Nilai classifier yang di peroleh yaitu 0,9 atau tingkat akurasinya 90 %**

```python
classifier.score(X_test, y_test)
```

**Membuat nilai prediksi**

```python
y_pred = classifier.predict(X_test)
y_pred
```

```python
y_real = y_test
y_real
```

```python
from sklearn.metrics import confusion_matrix
```

**Membuat variabel Cm dan menampilkan nilai matrix dengan mengambil nilai dari nilai sebenarnya**

```python
cm = confusion_matrix(y_test, y_pred)
```

```python
cm
```

**Confusion Matrix**

Confusion matrix juga sering disebut error matrix. Pada dasarnya confusion matrix memberikan informasi perbandingan hasil klasifikasi yang dilakukan oleh sistem (model) dengan hasil klasifikasi sebenarnya. Confusion matrix berbentuk tabel matriks yang menggambarkan kinerja model klasifikasi pada serangkaian data uji yang nilai sebenarnya diketahui.
