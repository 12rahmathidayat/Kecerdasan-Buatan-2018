# **Support Vector Machine (SVM)**

Support vector machine (SVM) adalah suatu teknik untuk melakukan prediksi, baik dalam kasus klasifikasi maupun regresi.

**DATASET**


Dataset yang digunakan adalah data bunga iris. Data bunga iris berisikan 3 data spesies iris
1. Setosa
2. Versicolor
3. Virginica

Fitur bunga iris yang diamati adalah panjang dan lebar, kelopak dan tangkai bunga.

**Library**

Import library yang dibutuhkan dan juga menampilkan dataset yang diambil

```
import pandas as pd
import seaborn as sb
from sklearn.datasets import load_iris
iris=load_iris()
iris
```

Membuat data X dan menampilkan data X nya

```
X=pd.DataFrame(iris.data, columns=iris.feature_names)
X.head()
```

Membuat data y dan menampilkan data y nya

```
y=pd.DataFrame(iris.target, columns=['species'])
y.head()
```

```
print(X.shape)
print(y.shape)
```

Menggunakan SVM untuk mengklasifikasi data

```
from sklearn.svm import SVC
clf=SVC(gamma='scale')
clf.fit(X,y)
```

Membandingkan y predict

```
y_pred=clf.predict(X)
y_pred=pd.DataFrame(y_pred)
y_pred.head()
```

Akurasi kalsifikasi

```
from sklearn.metrics import accuracy_score
accuracy_score(y, y_pred)
```

```
from sklearn.metrics import confusion_matrix
confusion_matrix(y, y_pred)
```

```
sb.set(font_scale=2)
sb.heatmap(confusion_matrix(y, y_pred), annot=True, fmt='d', cmap='viridis')
```
