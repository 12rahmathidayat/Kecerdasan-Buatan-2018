```python
import numpy as np
import pandas as pd
from sklearn import preprocessing
from sklearn.model_selection import train_test_split
from keras.models import Sequential
from keras.layers import Dense
```

```python
class MultiColumnLabelEncoder:
    def __init__(self,columns = None):
        self.columns = columns # array berisi nama kolom yang ingin diencode
        self.le = preprocessing.LabelEncoder()

    def fit_transform(self,X):
        output = X.copy()
        for col in self.columns:
            output[col] = self.le.fit_transform(output[col])
        return output
```

```python
dataset = pd.read_csv("play_tennis.csv")
print ("===== Dataset =====\n")
print(dataset)
```

```python
# encode dataset tenis dengan label encoder
dataset = MultiColumnLabelEncoder(columns = ['outlook','temp','humidity','wind', 'play']).fit_transform(dataset)
# split dataset secara acak jadi data training dan data tes dengan holdout split 10%
dataset_train, dataset_test = train_test_split(dataset, test_size=0.1)
dataset_train_labels = dataset_train['play'].values
dataset_train_data = dataset_train.drop(columns=['play']).values
dataset_test_labels = dataset_test['play'].values
dataset_test_data = dataset_test.drop(columns=['play']).values
```

```python
# Implementasi keras dengan menggunakan model sequential dan layer dense
model = Sequential()
model.add(Dense(2, input_dim=4, activation='sigmoid'))
model.add(Dense(3, activation='sigmoid'))
model.add(Dense(1, activation='sigmoid'))
# Compile model
# loss menggunakan binary_crossentropy karena loss function ini cocok diterapkan 
#  pada dataset dengan label biner, 0 dan 1.
# Optimizer menggunakan stochastic gradient descent SGD
model.compile(loss='binary_crossentropy', optimizer='SGD', metrics=['accuracy'])
# Fit model dengan jumlah epoch 150 dan batch size 1 atau stochastic
model.fit(dataset_train_data, dataset_train_labels, epochs=150, batch_size=1)
```

```python
# Evaluasi model terhadap data test
scores = model.evaluate(dataset_test_data, dataset_test_labels)
print("\n%s: %.2f%%" % (model.metrics_names[1], scores[1]*100))
```
